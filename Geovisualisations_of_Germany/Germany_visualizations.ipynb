```python
import geopandas as gpd
import descartes
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# importing an ESRI Shapefile and plotting it using GeoPandas
districtsDEU = gpd.read_file('data/DEU_adm/DEU_adm0.shp')
districtsDEU
```

```python
districtsDEU.plot(color = 'black', figsize = (15,13))
```

```python
DEUwat = gpd.read_file('data/DEU_wat/DEU_water_areas_dcw.shp')
DEUwat
```

```python
# plotting all water areas in Germany
DEUwat.plot(figsize = (15,13))
```

```python
DEUwatlines = gpd.read_file('data/DEU_wat/DEU_water_lines_dcw.shp')
DEUwatlines
```

```python
# plotting all rivers in Germany
DEUwatlines.plot(figsize = (15,13))
```

```python
DEUrds = gpd.read_file('data/DEU_rds/DEU_roads.shp')
DEUrds
```

```python
# plotting all roads in Germany
DEUrds.plot(figsize = (15,13), color = 'grey')
```

```python
# plotting multiple layers (Germany area and all rivers in Germany)
fig, ax = plt.subplots(figsize = (15, 13))
districtsDEU.plot(ax = ax, color = 'grey')
DEUwatlines.plot(ax = ax, color = 'blue')
DEUwat.plot(ax = ax, color = 'blue')
```

```python
# plotting multiple layers (Germany area and roads in Germany)
fig, ax = plt.subplots(figsize = (15, 13))
districtsDEU.plot(ax = ax, color = 'black')
DEUrds.plot(ax = ax, color = 'grey')
```
